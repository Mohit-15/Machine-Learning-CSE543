#Support Vector Regression using synthetic data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

Create a nonlinear function

```python
def nonlinear(array):
    return (10*array[:,0]-np.exp(0.01*array[:,1]+np.log(1+array[:,2]**2)))/(array[:,3]**2+5)
```

Generate features and target data for regression

```python
n_samples = 1000
n_features = 6
```

```python
x = 6*np.random.rand(n_samples,n_features)
x
```

```python
y = nonlinear(x)+np.random.randn(n_samples)
y
```

```python
y=y.reshape(n_samples,1)
```

```python
# Add code below
# read x and y arrays into a dataframe 'df_synthetic' , label the target as 'y' in the dataframe
df_synthetic = pd.DataFrame(x)
df_synthetic['y'] = y

df_synthetic.head()
```

Plot the data - Uncomment the following cell

```python
fig,ax = plt.subplots(2,3,figsize=(10,8))
ax = ax.ravel()
for i in range(6):
   ax[i].scatter(df_synthetic[df_synthetic.columns[i]],df_synthetic['y'],edgecolor='k',color='red',alpha=0.75)
   ax[i].set_title(f"{df_synthetic.columns[i]} vs. y",fontsize=14)
   ax[i].grid(True)
plt.show()
```

```python
# Add code below
from sklearn.model_selection import train_test_split
# Perform test train split of the dataframe 'df_synthetic'
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

print(f"{x_train.shape}, {x_test.shape}, {y_train.shape}, {y_test.shape}")
```

Support vector regressor with linear kernel

```python
# Add code below
from sklearn import svm
# Create linear support vector regressor model and fit it on the dataset
svr_model = svm.SVR(kernel ='linear')
svr_model.fit(x_train, y_train)
```

```python
# Add code below
from sklearn.metrics import r2_score
# Display the R2 score
svr_r2_score = r2_score(y_test, svr_model.predict(x_test))
print(f"R2 Score of SVR Regressor is: {svr_r2_score*100}")
```

```python
# Add code below
from sklearn.linear_model import LinearRegression
# Build a linear regressor on the above dataset and report its R2 score
linear_regressor_model = LinearRegression()
linear_regressor_model.fit(x_train, y_train)

linear_r2_score = r2_score(y_test, linear_regressor_model.predict(x_test))
print(f"R2 Score of Linear Regressor is: {linear_r2_score*100}")
```

Support vector regressor with RBF kernel

```python
#Add code below

# Use RBF kernel with default values and perform support vector regression
svr_rbf_model = svm.SVR(kernel ='rbf')
svr_rbf_model.fit(x_train, y_train)
```

```python
# Add code below
# Report R2 score of the above model
svr_rbf_r2_score = r2_score(y_test, svr_rbf_model.predict(x_test))
print(f"R2 Score of SVR Regressor with RBF Kernal is: {svr_rbf_r2_score*100}")
```

```python
# Add code below
from sklearn.model_selection import GridSearchCV
# Perform grid search of hyperparameters - cost C and epsilon, keep gamma at 'scale' (with cross-validation) to see if the test/validation score be improved
gscv = GridSearchCV(svm.SVR(),  {'C': [1.0, 1.5, 2, 2.5], 'epsilon': [0.1, 0.2, 0.3],'gamma': ['scale'], 'kernel': ['rbf','linear']}, cv = 5, scoring = 'r2', verbose =1, n_jobs = -1)
gscv.fit(x_train, y_train)
```

```python
# Add code below
# Check best model by the grid search and display it
print(gscv.best_estimator_)
```

```python
# Add code below
# Fit that model to the data and check the R2 score
best_model = svm.SVR(C=2.5, epsilon=0.3, gamma='scale', kernel='rbf')
best_model.fit(x_train, y_train)

best_score = r2_score(y_test, best_model.predict(x_test))
print(f"R2 Score of SVR Model with best hyperparameters: {best_score*100}")
```

```python
# printing all the r2_scores of the models created above in the form of table

scores = {'Model': ['SVR (linear)', 'Linear Regressor', 'SVR (rbf)', 'GridSearchCV Model'],
        'CV-Score': [svr_r2_score*100, linear_r2_score*100, svr_rbf_r2_score*100, best_score*100]}

score_table = pd.DataFrame(scores)
score_table
```
